```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Reading the image
img=plt.imread("C:/Users/KIIT/Downloads/group_photo.jpg")
```

```python
#Display the image
plt.imshow(img)
```

```python
img.shape
```

```python
#using haar-cascade for feature extraction
model=cv2.CascadeClassifier("C:/Users/KIIT/Downloads/haarcascade_frontalface_alt.xml")
```

```python
#Detecting features
faces=model.detectMultiScale(img)
```

```python
faces
```

```python
faces.shape
```

```python
#Creating the rectangle for faces
for x,y,w,h in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),2)
```

```python
#Display the result
plt.imshow(img)
```
